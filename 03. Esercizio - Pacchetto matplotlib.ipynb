## Preparazione di un Grafico

**Innanzitutto va importato il sottopacchetto plt**

```python
from matplotlib import pyplot as plt
```

* Il sottopacchetto pyplot viene di solito rinominato come plot

**Quindi, si costruisce una nuova figura con figure**

```python
plt.figure(figsize=(20, 3))
```

* Si può specificare la dimensione con l'argomento figsize
* La dimensione è una tupla nel formato (larghezza, altezza)

## Funzione `plot`

**La funzione di disegno plot permette di disegnare curve**

Proveremo ad usarla per disegnare la funzione: $f(x) = \sin(x) + 0.1x$

* Per prima cosa, costruiamo due liste con le coordinate $x$ e $y$ di punti sulla curva

```python
import math

def linspace(l, u, n):
    return [l + (u - l) * v / (n-1) for v in range(n)]

x = linspace(0, 10, 12)
y = [math.sin(v) + 0.1*v for v in x]
n = 4
print(f'Primi {n} elementi di x: {x[:n]}')
print(f'Primi {n} elementi di y: {y[:n]}')
```

* `linspace` genera $n$ punti equispaziati nell'intervallo $[l, u]$

## Funzione `plot`

**Ora possiamo invocare plot(x, y, ...) per disegnare la curva**

* Gli argomenti `x` e `y` sono due collezioni con le coordinate di punti sulla curva
* La funzioen ottiene un disegno collegando con una retta i punti adiacenti

```python
plt.figure(figsize=(20, 4))
plt.plot(x, y)
plt.show()
```

## Funzione `plot`

**Aumentando il numero di punti per unità possiamo migliorare il disegno**

```python
x = linspace(0, 10, 100)
y = [math.sin(v) + 0.1*v for v in x]

plt.figure(figsize=(20, 4))
plt.plot(x, y)
plt.show()
```

## Funzione `plot`

**Si può disegnare più di una curva sullo stesso grafico**

```python
x = linspace(0, 10, 100)
y = [math.sin(v) + 0.1*v for v in x]
y2 = [math.cos(v) + 0.1*v for v in x]

plt.figure(figsize=(20, 4))
plt.plot(x, y)
plt.plot(x, y2) # Disegno la seconda curva
plt.show()
```

## Esercizio

**Provare a graficare la funzione $1.5x^3 - 1.5x^2 -x -0.4$**

* Disegnatela nell'intervallo $[-1, 1.5]$, con diverse densità di punti

```python
x = linspace(-1, 1.5, n=100)
y = [1.5*v**3 - 1.5*v**2 - v - 0.4 for v in x]

plt.figure(figsize=(20, 4))
plt.plot(x, y)
plt.show()
```

## Esercizio

**Confrontate sull'intervallo discreto $\{0, 1, \ldots 6\}$

...Le funzioni $x!$ ed $e^x$

* Per $x!$ si usi la funzione `math.factorial`
* L'intervallo discreto può essere ottenuto convertendo in lista un `range`

```python
x = list(range(0, 7))
y = [math.exp(v) for v in x]
y2 = [math.factorial(v) for v in x]

plt.figure(figsize=(20, 4))
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

## Esercizio

**Graficate sull'intervallo $[-5, 5]$ la seguente funzione definita per casi**

$$
f(x) = \left\{\begin{aligned}
& e^{-\frac{1}{x^2}} & \text{if $x \neq 0$} \\
& 0 & \text{otherwise}
\end{aligned}\right.
$$

```python
x = linspace(-5, 5, 100)
y = [math.exp(-1/v**2) if v != 0 else 0 for v in x]
plt.figure(figsize=(20, 4))
plt.plot(x, y)
plt.show()
```
